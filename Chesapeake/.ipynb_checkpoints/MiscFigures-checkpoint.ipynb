# Get slice data for each case:

```python
cases_of_interest = [
                     'ERAI_YSU_CHRN_OST_NOSK_4DOM',
                     'ERAI_YSU_CHRN_OST_OCEA_3DOM','ERAI_YSU_CHRN_GHR_OCEA_3DOM',
                     'ERAI_YSU_JMNZ_OST_NOSK_3DOM',
                     'ERAI_YSU_CHRN_MOD_OCEA_3DOM',
                     'GFSR_YSU_CHRN_GFS_NOSK_3DOM',
                     'GFSR_YSU_JMNZ_GFS_NOSK_3DOM',
                     'ERA5_YSU_CHRN_ER5_NOSK_3DOM',
                     'ERA5_YSU_JMNZ_ER5_NOSK_3DOM',
                     'ERAI_YSU_CHRN_GHR_NOSK_3DOM','ERAI_YSU_CHRN_MOD_NOSK_3DOM'
                    ]
```

```python
wrfin = xr.open_dataset('{}{}/{}/wrfinput_d03'.format(wrf_dir,cases_of_interest[1],restarts[0]))

landmask = np.squeeze(wrfin.LANDMASK)
lat = np.squeeze(wrfin.XLAT)
lon = np.squeeze(wrfin.XLONG)
water_dpth = np.squeeze(wrfin.WATER_DEPTH)

wrfin.close()

avg_cases = {}

for cc,case in enumerate(cases_of_interest):
    file_paths = []
    dom = int(str(case_dict[case].keys()).split("'")[1].replace('d0',''))
    new_f_name = '{0}{1}/{1}_d0{2}_average_data.nc'.format(wrf_dir,case,dom)
    f_dir = '{}{}/{}/'.format(wrf_dir,case,'{}')
    if path.exists(new_f_name):
        print('loading in full dataset!')
        avg_cases[case] = xr.open_dataset(new_f_name)
    else:
        for rr,rst in enumerate(restarts[:1]):
            f_list = sorted(glob.glob('{}wrfout_d0{}*'.format(f_dir.format(rst),dom)))
            if rr < len(restarts)-1:
                f_list_good = f_list[n_spinup_files:-1]
            else:
                f_list_good = f_list[n_spinup_files:]        
            for ff in f_list_good: 
                file_paths.append(ff)

        init_vars = True
        for ff,filen in enumerate(file_paths):
            wrf_f = xr.open_dataset(filen)

            z0  = wrf_f.ZNT
            ust = wrf_f.UST
            u10 = wrf_f.U10
            v10 = wrf_f.V10
            t2  = wrf_f.T2
            wrf_f.close()

            if init_vars:
                z0f  = z0.copy()
                ustf = ust.copy()
                u10f = u10.copy()
                v10f = v10.copy()
                t2f  = t2.copy()
                init_vars = False
            else:
                z0f  = xr.concat([z0f,z0],dim='Time')
                ustf = xr.concat([ustf,ust],dim='Time')
                u10f = xr.concat([u10f,u10],dim='Time')
                v10f = xr.concat([v10f,v10],dim='Time')
                t2f  = xr.concat([t2f,t2],dim='Time')
        z0m  = z0f.mean(dim='Time')
        ustm = ustf.mean(dim='Time')
        u10m = u10f.mean(dim='Time')
        v10m = v10f.mean(dim='Time')
        t2m  = t2f.mean(dim='Time')
        ds = xr.merge([z0m,t2m,ustm,u10m,v10m,landmask,water_dpth])
        ds = ds.assign_coords({'lat':lat,'lon':lon})
        ds = ds.drop(['lat','lon'])
        ds.to_netcdf(new_f_name)
        avg_cases[case] = ds
        print('Finished with {}'.format(case))
```

```python
diff_controls  = ['ERAI_YSU_CHRN_OST_NOSK_4DOM','GFSR_YSU_CHRN_GFS_NOSK_3DOM','ERA5_YSU_CHRN_ER5_NOSK_3DOM']
diff_jmnz      = ['ERAI_YSU_JMNZ_OST_NOSK_3DOM','GFSR_YSU_JMNZ_GFS_NOSK_3DOM','ERA5_YSU_JMNZ_ER5_NOSK_3DOM']
ocean_controls = ['ERAI_YSU_CHRN_OST_NOSK_4DOM','ERAI_YSU_CHRN_GHR_NOSK_3DOM','ERAI_YSU_CHRN_MOD_NOSK_3DOM']
ocean_oml      = ['ERAI_YSU_CHRN_OST_OCEA_3DOM','ERAI_YSU_CHRN_GHR_OCEA_3DOM','ERAI_YSU_CHRN_MOD_OCEA_3DOM']
num_diffs = len(diff_controls)
```

```python
fig,ax = plt.subplots(nrows=4,ncols=3,figsize=(18,18),sharey=True,sharex=True)

for cc in range(0,num_diffs):
    ds_cntrl = avg_cases[diff_controls[cc]]
    ds_jmnz  = avg_cases[diff_jmnz[cc]]
    znt_diff = ds_jmnz.ZNT - ds_cntrl.ZNT
    ust_diff = ds_jmnz.UST - ds_cntrl.UST
    vel_diff = (ds_jmnz.U10**2 + ds_jmnz.V10**2)**0.5 - (ds_cntrl.U10**2 + ds_cntrl.V10**2)**0.5
    t2_diff  = ds_jmnz.T2 - ds_cntrl.T2
    
    txt_x,txt_y = -80.0,41.35
    water_levels = np.arange(10.0,101.0,20.0)
    
    znt_plt = ax[0,cc].pcolormesh(ds_cntrl.XLONG, ds_cntrl.XLAT,znt_diff.where(ds_jmnz.LANDMASK == 0.0), 
                       cmap=plt.cm.coolwarm, norm=Normalize(-1.5e-3,1.5e-3))
    ax[0,cc].contour(ds_cntrl.XLONG, ds_cntrl.XLAT, ds_cntrl.LANDMASK,levels=[0.5],colors='k',alpha=1.0,linewidths=1.0)
    ax[0,cc].contour(ds_cntrl.XLONG, ds_cntrl.XLAT, ds_cntrl.WATER_DEPTH,levels=water_levels,colors='k',alpha=0.5,linestyles='-')
    ax[0,cc].tick_params(labelsize=15)
    ax[0,cc].text(txt_x,txt_y,'z$_0$',size=20,ha='left',va='top')
    
    
    ust_plt = ax[1,cc].pcolormesh(ds_cntrl.XLONG, ds_cntrl.XLAT, ust_diff.where(ds_jmnz.LANDMASK == 0.0), 
                       cmap=plt.cm.coolwarm, norm=Normalize(-0.07,0.07))
    ax[1,cc].contour(ds_cntrl.XLONG, ds_cntrl.XLAT, ds_cntrl.LANDMASK,levels=[0.5],colors='k',alpha=1.0,linewidths=1.0)
    ax[1,cc].contour(ds_cntrl.XLONG, ds_cntrl.XLAT, ds_cntrl.WATER_DEPTH,levels=water_levels,colors='k',alpha=0.5)
    ax[1,cc].tick_params(labelsize=15)
    ax[1,cc].text(txt_x,txt_y,'u$_*$',size=20,ha='left',va='top')
    
    
    vel_plt = ax[2,cc].pcolormesh(ds_cntrl.XLONG, ds_cntrl.XLAT, vel_diff.where(ds_jmnz.LANDMASK == 0.0), 
                       cmap=plt.cm.coolwarm, norm=Normalize(-0.7,0.7))
    ax[2,cc].contour(ds_cntrl.XLONG, ds_cntrl.XLAT, ds_cntrl.LANDMASK,levels=[0.5],colors='k',alpha=1.0,linewidths=1.0)
    ax[2,cc].contour(ds_cntrl.XLONG, ds_cntrl.XLAT, ds_cntrl.WATER_DEPTH,levels=water_levels,colors='k',alpha=0.5)
    ax[2,cc].tick_params(labelsize=15)
    ax[2,cc].text(txt_x,txt_y,'U at 10 m',size=20,ha='left',va='top')
    
    
    t2_plt = ax[3,cc].pcolormesh(ds_cntrl.XLONG, ds_cntrl.XLAT, t2_diff.where(ds_jmnz.LANDMASK == 0.0),
                       cmap=plt.cm.coolwarm, norm=Normalize(-0.3,0.3))
    ax[3,cc].contour(ds_cntrl.XLONG, ds_cntrl.XLAT, ds_cntrl.LANDMASK,levels=[0.5],colors='k',alpha=1.0,linewidths=1.0)
    ax[3,cc].contour(ds_cntrl.XLONG, ds_cntrl.XLAT, ds_cntrl.WATER_DEPTH,levels=water_levels,colors='k',alpha=0.5)
    ax[3,cc].tick_params(labelsize=15)
    ax[3,cc].text(txt_x,txt_y,'T2',size=20,ha='left',va='top')
    ax[3,cc].set_xlabel('Longitude',size=15)
    
ax[0,0].set_ylabel('Latitude',size=15)
ax[1,0].set_ylabel('Latitude',size=15)
ax[2,0].set_ylabel('Latitude',size=15)
ax[3,0].set_ylabel('Latitude',size=15)

ax[0,0].set_title('ERA-Interim',size=22)
ax[0,1].set_title('GFS',size=22)
ax[0,2].set_title('ERA5',size=22)

znt_cbar = fig.colorbar(znt_plt, ax=ax[0, :])
znt_cbar.ax.tick_params(labelsize=14)
ust_cbar = fig.colorbar(ust_plt, ax=ax[1, :])
ust_cbar.ax.tick_params(labelsize=14)
vel_cbar = fig.colorbar(vel_plt, ax=ax[2, :])
vel_cbar.ax.tick_params(labelsize=14)
t2_cbar  = fig.colorbar(t2_plt,  ax=ax[3, :])
t2_cbar.ax.tick_params(labelsize=14)

plt.show()
                    
```
